```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the training dataset
train_df = pd.read_csv("/content/Titanic_train.csv")

# Display the first 5 rows, column names, and data types of the training dataset
print("Training Data - First 5 Rows:")
display(train_df.head())
print("\nTraining Data - Column Names and Data Types:")
display(train_df.info())

# Load the testing dataset
test_df = pd.read_csv("/content/Titanic_test.csv")

# Display the first 5 rows, column names, and data types of the testing dataset
print("\nTesting Data - First 5 Rows:")
display(test_df.head())
print("\nTesting Data - Column Names and Data Types:")
display(test_df.info())

# Display summary statistics for numerical columns in the training dataset
print("\nTraining Data - Summary Statistics:")
display(train_df.describe())

# Display the number of missing values in each column of the training dataset
print("\nTraining Data - Missing Values:")
display(train_df.isnull().sum())

# Display the number of unique values in each column of the training dataset
print("\nTraining Data - Unique Values:")
display(train_df.nunique())
```

```python
# Create a histogram for the 'Age' column
plt.figure(figsize=(10, 6))
sns.histplot(train_df['Age'].dropna(), kde=True, bins=30)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Create a box plot for the 'Fare' column
plt.figure(figsize=(10, 6))
sns.boxplot(x=train_df['Fare'])
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.show()

# Create a count plot for the 'Survived' column
plt.figure(figsize=(8, 5))
sns.countplot(x='Survived', data=train_df)
plt.title('Count of Survival (0 = No, 1 = Yes)')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Impute missing 'Age' values with the median of the training data
age_median = train_df['Age'].median()
train_df['Age'].fillna(age_median, inplace=True)
test_df['Age'].fillna(age_median, inplace=True)

# Impute missing 'Embarked' values with the mode of the training data
embarked_mode = train_df['Embarked'].mode()[0]
train_df['Embarked'].fillna(embarked_mode, inplace=True)

# Impute missing 'Fare' value with the median of the training data
fare_median = train_df['Fare'].median()
test_df['Fare'].fillna(fare_median, inplace=True)

# Drop the 'Cabin' column
train_df.drop('Cabin', axis=1, inplace=True)
test_df.drop('Cabin', axis=1, inplace=True)

# One-hot encode 'Sex' and 'Embarked' columns
train_df = pd.get_dummies(train_df, columns=['Sex', 'Embarked'], drop_first=True)
test_df = pd.get_dummies(test_df, columns=['Sex', 'Embarked'], drop_first=True)

# Drop unnecessary columns
train_df.drop(['Name', 'Ticket'], axis=1, inplace=True)
test_df.drop(['Name', 'Ticket'], axis=1, inplace=True)

# Align columns - crucial for consistent feature sets
train_labels = train_df['Survived']
train_ids = train_df['PassengerId']
test_ids = test_df['PassengerId']

# Align columns after one-hot encoding, before splitting data
# This ensures both dataframes have the same dummy variables
train_df, test_df = train_df.align(test_df, join='inner', axis=1)

print("Data after cleaning and encoding:")
print("\nTraining Data:")
display(train_df.head())
print("\nTesting Data:")
display(test_df.head())
```

```python
from sklearn.linear_model import LogisticRegression

# Instantiate the model
logreg = LogisticRegression(random_state=42)

# Train the model
logreg.fit(train_df, train_labels)

print("Logistic Regression model trained successfully.")
```

```python
from sklearn.preprocessing import StandardScaler

# Scale the data
scaler = StandardScaler()
train_df_scaled = scaler.fit_transform(train_df)
test_df_scaled = scaler.transform(test_df)

# Re-instantiate and retrain the model with more iterations
logreg = LogisticRegression(random_state=42, max_iter=1000)
logreg.fit(train_df_scaled, train_labels)

print("Logistic Regression model retrained successfully with scaled data and increased iterations.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Predict probabilities on the training data
train_pred_proba = logreg.predict_proba(train_df_scaled)[:, 1]

# Generate class predictions on the training data
train_pred = logreg.predict(train_df_scaled)

# Calculate and print evaluation metrics
accuracy = accuracy_score(train_labels, train_pred)
precision = precision_score(train_labels, train_pred)
recall = recall_score(train_labels, train_pred)
f1 = f1_score(train_labels, train_pred)
roc_auc = roc_auc_score(train_labels, train_pred_proba)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
print(f"ROC-AUC score: {roc_auc:.4f}")

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(train_labels, train_pred_proba)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--', label='Random classifier')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
# Create a DataFrame of feature coefficients
coefficients_df = pd.DataFrame({
    'Feature': train_df.columns,
    'Coefficient': logreg.coef_[0]
})

# Sort by the absolute value of the coefficients in descending order
coefficients_df['Abs_Coefficient'] = coefficients_df['Coefficient'].abs()
coefficients_df = coefficients_df.sort_values(by='Abs_Coefficient', ascending=False).drop(columns='Abs_Coefficient')

# Print the sorted DataFrame
print("Logistic Regression Coefficients (sorted by influence):")
display(coefficients_df)

# Interpretation of the coefficients
print("\n--- Interpretation of Coefficients ---")
print("The coefficients in a logistic regression model represent the change in the log-odds of the outcome (survival) for a one-unit change in the predictor feature.")
print("\n- **Positive Coefficient**: A positive coefficient indicates that an increase in the feature's value increases the log-odds of survival, making survival more likely. For example, a positive coefficient for 'Sex_male' would suggest that being male increases the chances of survival.")
print("\n- **Negative Coefficient**: A negative coefficient means that an increase in the feature's value decreases the log-odds of survival, making survival less likely. For instance, a negative coefficient for 'Pclass' suggests that being in a lower class (higher 'Pclass' number) decreases the chances of survival.")
print("\nFrom the sorted table, we can see that 'Sex_male' has the largest absolute coefficient, indicating it's the most significant predictor of survival. Other important features include 'Pclass' and 'Age'.")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import pickle
from sklearn.preprocessing import StandardScaler

# Load the trained model and scaler
try:
    with open('logistic_regression_model.pkl', 'rb') as model_file:
        logreg = pickle.load(model_file)
    with open('scaler.pkl', 'rb') as scaler_file:
        scaler = pickle.load(scaler_file)
except FileNotFoundError:
    st.error("Model or scaler file not found. Please ensure 'logistic_regression_model.pkl' and 'scaler.pkl' are in the same directory.")
    st.stop()

# Define the Streamlit app layout
st.title('Titanic Survival Prediction')

st.write("""
Enter the passenger's details to predict their survival probability.
""")

# Create input widgets for features
pclass = st.selectbox('Passenger Class (1st, 2nd, 3rd)', [1, 2, 3])
age = st.slider('Age', 0.42, 80.0, 28.0) # Using min/max from describe, median for default
sibsp = st.slider('Number of Siblings/Spouses Aboard', 0, 8, 0) # Using max from describe
parch = st.slider('Number of Parents/Children Aboard', 0, 6, 0) # Using max from describe
fare = st.slider('Fare', 0.0, 512.3292, 14.4542) # Using min/max from describe, median for default
sex = st.selectbox('Sex', ['male', 'female'])
embarked = st.selectbox('Port of Embarkation', ['C', 'Q', 'S'])

# Map categorical inputs to model's dummy variable format
sex_male = True if sex == 'male' else False
embarked_Q = True if embarked == 'Q' else False
embarked_S = True if embarked == 'S' else False

# Create a button to trigger prediction
if st.button('Predict Survival'):
    # Create a DataFrame from user inputs
    # Ensure column order matches the training data order:
    # 'PassengerId', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_male', 'Embarked_Q', 'Embarked_S'
    # We will use dummy PassengerId for prediction input
    input_data = pd.DataFrame([[9999, pclass, age, sibsp, parch, fare, sex_male, embarked_Q, embarked_S]],
                                columns=['PassengerId', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_male', 'Embarked_Q', 'Embarked_S'])

    # Scale the input data
    # Need to handle potential future columns if model was trained on more
    # For simplicity, let's assume the scaler fits the input columns
    try:
        input_data_scaled = scaler.transform(input_data)
    except ValueError as e:
        st.error(f"Error scaling input data. Ensure input features match scaler features: {e}")
        st.stop()


    # Make prediction
    prediction_proba = logreg.predict_proba(input_data_scaled)[:, 1]
    prediction = logreg.predict(input_data_scaled)[0]

    # Display the prediction result
    st.subheader('Prediction Result:')
    if prediction == 1:
        st.success(f'Prediction: Survived (Probability: {prediction_proba[0]:.2f})')
    else:
        st.error(f'Prediction: Did not survive (Probability: {prediction_proba[0]:.2f})')

st.write("""
**Instructions to run the app:**
1. Save the trained model and scaler objects as 'logistic_regression_model.pkl' and 'scaler.pkl' in the same directory as this 'app.py' file.
2. Open your terminal or command prompt.
3. Navigate to the directory where 'app.py' is saved.
4. Run the command: `streamlit run app.py`
5. The app will open in your web browser.
""")
```

```python
import pickle

# Save the logistic regression model
with open('logistic_regression_model.pkl', 'wb') as model_file:
    pickle.dump(logreg, model_file)

# Save the scaler object
with open('scaler.pkl', 'wb') as scaler_file:
    pickle.dump(scaler, scaler_file)

print("Model and scaler have been saved as 'logistic_regression_model.pkl' and 'scaler.pkl'")
```
